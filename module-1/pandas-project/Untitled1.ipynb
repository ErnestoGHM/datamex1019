```python
import numpy as np 
```

```python
boys = np.random.normal(70,5,1000)
girls = np.random.normal(50,3,1000)
```

```python
for i in range(100):
    girls[np.random.randint(0,1000)] = np.nan
```

```python
import pandas as pd
```

```python
boys = pd.DataFrame(boys,columns=['weight'])
boys['gender']='boy'

girls = pd.DataFrame(girls,columns=['weight'])
girls['gender']='girl'

display(boys)
display(girls)
```

```python
df = pd.concat([girls,boys],axis=0)
df['weight']=df['weight'].astype(float)
```

```python
df
```

```python
df['filled_weight']=df.groupby('gender')['weight'].transform(lambda grp:grp.fillna(np.mean(grp)))
df['filled_weight']
```
